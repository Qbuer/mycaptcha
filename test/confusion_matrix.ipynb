```python
import os
```

```python
dir = "../debug"
files = os.listdir(dir)
golds, preds = [], []
for file in files:
    gold, pred = file.split('.')[0].split('-')
    for g, p in zip(gold, pred):
        golds.append(g)
        preds.append(p)
labels = list(set(golds))
labels.sort()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python

cm = confusion_matrix(golds, preds, labels=labels)
disp = ConfusionMatrixDisplay(cm, display_labels=labels)
disp.plot()
plt.show()
```

```python
with open('confusion_matrix.tsv', 'w') as writer:
    
    writer.write("-\t" + "\t".join(labels))
    writer.write("\n")
    for row, label in zip(cm.tolist(), labels):

        writer.write(label+"\t" + "\t".join([str(x) for x in row]))
        writer.write("\n")
```
