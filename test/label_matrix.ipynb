```python
import os
```

```python
dir = "../data/dataset1/train"
files = os.listdir(dir)
from collections import Counter
counter = Counter()
for file in files:
    for char in file.split('.')[0]:
        counter[char] += 1
```

```python
import matplotlib.pyplot as plt
key, value = [], []
for k, v in counter.items():
    key.append(k)
    value.append(v)
plt.bar(key, value)
plt.show()
```
